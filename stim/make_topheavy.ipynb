```python
from PIL import Image
import numpy as np
from IPython.display import display # to display images

%matplotlib inline

import matplotlib.pyplot as plt
#import matplotlib.image as mpimg
import matplotlib.patches as mpatches

import random
```

```python
#Placement parameters
cent =568 #center of face
y_max = 180 #How far up a dot can be placed
x_max =180 #How far left a dot can be placed
max_circ = 10
circ_size = 30
spacing = 10
prop = 1/4 #propotion the go down
num_img = 60

im = plt.imread('face_template.png') #load template
```

```python

for kk in range(1, num_img+1):
    num_circ = random.randint(3,max_circ)
    if random.random() > .5:
        down_circ = round(num_circ * prop)
    else:
        down_circ = 0
    # Create a figure. Equal aspect so circles look circular
    fig,ax = plt.subplots(1)
    ax.set_aspect('equal')

    # Show the image
    ax.imshow(im)

    for ii in range(0, num_circ+1):
        #if the last circle, make it go on bottom
        pos = random.random()
        if ii <= (num_circ-down_circ): 
            vert_loc = 1 
            #print(pos, vert_loc)
        else: 
            vert_loc = -1
        
            
        x_coord = cent - random.randint(0, x_max)
        y_coord = cent -  vert_loc * (random.randint(60,y_max))
        #y_coord = cent
        
        #print(cent, vert_loc, x_coord, y_coord, ii, num_circ)

        #Add left side circles
        circ = mpatches.Circle((x_coord,y_coord), circ_size, fc = 'black', ec ='black')
        ax.add_patch(circ)

        #mirror to right side
        x_coord =(cent-x_coord)*2 + x_coord
        circ = mpatches.Circle((x_coord,y_coord), circ_size, fc = 'black', ec ='black')
        ax.add_patch(circ)

    #Save Fig
    plt.axis('off')
    plt.savefig(f'Schematic/schematic_{kk}.png', bbox_inches='tight', pad_inches = 0, dpi=300)
    plt.show()

```
